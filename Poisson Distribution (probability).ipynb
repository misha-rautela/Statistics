The Poisson Distribution is a probability distribution that represents the number of events occurring within a fixed interval of time or space, given that the events occur with a known constant rate and independently of the time since the last event.

Call center use case: on average, the center receives 5 calls per hour. We want to calculate the probability of receiving exactly 3 calls in an hour

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameter for the Poisson distribution (average number of calls per hour)
lambda_ = 5

# Create an array of possible values for the number of calls (k = 0 to 15)
k_values = np.arange(0, 16)

# Calculate the Poisson probability mass function (PMF) for each k
pmf_values = stats.poisson.pmf(k_values, lambda_)

# Plotting the Poisson distribution
plt.bar(k_values, pmf_values, color='blue', alpha=0.7)
plt.title(f'Poisson Distribution (Î»={lambda_})')
plt.xlabel('Number of Calls (k)')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
# Calculate the probability of getting exactly 3 calls
k = 3
probability_3_calls = stats.poisson.pmf(k, lambda_)

print(f"Probability of receiving exactly 3 calls in an hour: {probability_3_calls:.4f}")
```
