The Normal Distribution (also known as the Gaussian Distribution) is one of the most common probability distributions in statistics. It is symmetric and describes data that clusters around the mean.

Use Case:
Analyzing the height of a group of people, which follows a Normal Distribution. We want to:

Visualize the normal distribution curve.

Calculate the probability of a value (e.g., probability of a person having a height between 170 cm and 180 cm).

Calculate the z-score, which tells us how many standard deviations a data point is from the mean.

We'll assume that the heights of the people in the group follow a Normal Distribution with:

Mean (µ): 175 cm

Standard Deviation (σ): 10 cm

Step 1: Visualizing the Normal Distribution
We can use matplotlib and scipy.stats to generate a normal distribution curve.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters for the normal distribution
mean = 175    # Mean height in cm
std_dev = 10   # Standard deviation in cm

# Create a range of heights from 140 to 210
heights = np.linspace(140, 210, 1000)

# Calculate the probability density function (PDF) for each height value
pdf_values = stats.norm.pdf(heights, mean, std_dev)

# Plotting the normal distribution curve
plt.plot(heights, pdf_values, label=f"Normal Distribution\nMean = {mean}, Std = {std_dev}")
plt.fill_between(heights, pdf_values, alpha=0.3)
plt.title('Normal Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

Step 2: calculate the cumulative probability for 170 cm and 180 cm and then find the difference between these two cumulative probabilities.

```python
# Calculate the cumulative probability (CDF) for heights 170 and 180
prob_170 = stats.norm.cdf(170, mean, std_dev)
prob_180 = stats.norm.cdf(180, mean, std_dev)

# The probability of height being between 170 and 180
probability_170_180 = prob_180 - prob_170

print(f"Probability of height being between 170 cm and 180 cm: {probability_170_180:.4f}")
```

```python
# Calculate the z-score for height 180 cm
height = 180
z_score = (height - mean) / std_dev

print(f"Z-score for height {height} cm: {z_score:.2f}")
```

The z-score for a height of 180 cm is 0.50, meaning the height of 180 cm is 0.5 standard deviations above the mean of 175 cm.
