```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Example Data: House Size (in square feet) and Price (in $)
house_size = np.array([800, 1000, 1200, 1400, 1600, 1800, 2000]).reshape(-1, 1)  # Independent variable (x)
house_price = np.array([150000, 180000, 210000, 250000, 270000, 300000, 330000])  # Dependent variable (y)

# Step 3: Create a Linear Regression model
model = LinearRegression()

# Step 4: Train the model
model.fit(house_size, house_price)

# Step 5: Make Predictions
predicted_price = model.predict(house_size)

# Step 6: Evaluate the Model
mse = mean_squared_error(house_price, predicted_price)
r2 = r2_score(house_price, predicted_price)

# Output the results
print(f"Intercept (β₀): {model.intercept_}")
print(f"Slope (β₁): {model.coef_[0]}")
print(f"Mean Squared Error: {mse}")
print(f"R² (Coefficient of Determination): {r2}")

# Step 7: Visualizing the Data and the Regression Line
plt.scatter(house_size, house_price, color='blue', label='Actual Data')
plt.plot(house_size, predicted_price, color='red', label='Regression Line')
plt.title('Linear Regression: House Price vs. House Size')
plt.xlabel('House Size (sq. feet)')
plt.ylabel('Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
